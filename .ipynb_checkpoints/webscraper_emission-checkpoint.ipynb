```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pprint import pprint
url = "https://pl.wikipedia.org/wiki/Lista_pa%C5%84stw_wed%C5%82ug_rocznej_emisji_dwutlenku_w%C4%99gla"
headers_req = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
}

response = requests.get(url, headers=headers_req)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find("table", {"class": "wikitable"})
headers = [th.get_text(strip=True) for th in table.find("tr").find_all("th")]
```

```python
data = []
for row in table.find_all("tr")[1:]:  
    cells = row.find_all("td")
    if cells: 
        data.append([cell.get_text(strip=True) for cell in cells])


df = pd.DataFrame(data, columns=headers[:len(data[0])])
```

```python
df = df.replace(",", ".", regex=True)
```

```python
for col in df.columns:
    if col=="Kraj":
        df[col] = df[col].astype(str)
    else:
        df[col] = pd.to_numeric(df[col], errors="coerce")
cols = df.columns.difference(['Kraj'])
df.loc[30, cols] = df.loc[30, cols].fillna(df.loc[30, cols].median())
print(df.head())
polish_to_english = {'Chiny':'China', 'Stany Zjednoczone':'United States', 'Indie':'India', 'Rosja':'Russia', 'Japonia':'Japan', 'Niemcy':'Germany', 'Korea Południowa':'South Korea', 'Arabia Saudyjska':'Saudi Arabia', 'Iran':'Iran', 'Kanada':'Canada', 'Brazylia':'Brazil', 'Indonezja':'Indonesia', 'Meksyk':'Mexico', 'Wielka Brytania':'United Kingdom', 'Południowa Afryka':'South Africa', 'Australia':'Australia', 'Turcja':'Turkey', 'Francja':'France', 'Włochy':'Italy', 'Tajlandia':'Thailand', 'Tajwan':'Taiwan', 'Polska':'Poland', 'Hiszpania':'Spain', 'Zjednoczone Emiraty Arabskie':'United Arab Emirates', 'Malezja':'Malaysia', 'Ukraina':'Ukraine', 'Singapur':'Singapore', 'Holandia':'Netherlands', 'Egipt':'Egypt', 'Argentyna':'Argentina', 'Kazachstan':'Kazakhstan', 'Wenezuela':'Venezuela', 'Pakistan':'Pakistan', 'Wietnam':'Vietnam', 'Belgia':'Belgium', 'Algieria':'Algeria', 'Katar':'Qatar', 'Uzbekistan':'Uzbekistan', 'Czechy':'Czech Republic', 'Kuwejt':'Kuwait', 'Chile':'Chile', 'Rumunia':'Romania', 'Grecja':'Greece'}
df['Kraj'] = df['Kraj'].replace(polish_to_english)
```
